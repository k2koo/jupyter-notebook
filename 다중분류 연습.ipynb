```python
import pandas as pd
data=pd.read_csv('Fvote.csv', encoding='utf-8')
```

```python
x=data[data.columns[1:13]]
y=data[['parties']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, stratify=y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
pred_train=model.predict(x_train)
model.score(x_train, y_train)
```

```python
pred_test=model.predict(x_test)
model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_train=confusion_matrix(y_train, pred_train)
print(confusion_train)
```

```python
confusion_test=confusion_matrix(y_test, pred_test)
print(confusion_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C': [0.001, 0.01, 0.1, 1, 10, 100]}
```

```python
grid_search=GridSearchCV(LogisticRegression(), param_grid, cv=5,
                        return_train_score=True)
grid_search.fit(x_train, y_train)
```

```python
print("베스트: {}".format(grid_search.best_params_))
print("교차: {:.44f}".format(grid_search.best_score_))
```

```python
print("테스트:{:.3f}".format(grid_search.score(x_test,y_test)))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param_distribs={'C': randint(low=0.001, high=100)}
```

```python
random_search=RandomizedSearchCV(LogisticRegression(),
                                param_distributions=param_distribs, cv=5,
                                # n_iter=100,랜덤횟수 디폴트=10
                                return_train_score=True)
random_search.fit(x_train, y_train)
```

```python
print("베스트:ㅖ{}".format(random_search.best_params_))
print("베스트크로스:{:.3f}".format(random_search.best_score_))
```

```python
print("테스트:{:.3f}".format(random_search.score(x_test, y_test)))
```
